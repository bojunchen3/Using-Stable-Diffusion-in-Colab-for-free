```python
import os
import ipywidgets as widgets
from IPython.display import display, clear_output
from PIL import Image
from io import BytesIO
from google.colab import files

save_counter = {"count": 0}

def interactive_image_downloader(images, save_prefix="image", max_width=300):
    image_boxes = []

    for i, img in enumerate(images):
        buf = BytesIO()
        img.save(buf, format="PNG")
        img_widget = widgets.Image(value=buf.getvalue(), format='png', width=max_width)

        download_button = widgets.Button(description="Download", button_style="info", layout=widgets.Layout(width="80px"))
        output = widgets.Output()

        def make_download_handler(img=img, idx=i):
            def handler(b):
                output.clear_output()
                filename = f"{save_prefix}_{save_counter['count']}.png"
                img.save(filename)
                files.download(filename)
                save_counter["count"] += 1
            return handler

        download_button.on_click(make_download_handler())

        image_boxes.append(
            widgets.VBox([img_widget, download_button, output])
        )

    grid = widgets.GridBox(
        children=image_boxes,
        layout=widgets.Layout(grid_template_columns="repeat(3, 33%)", grid_gap="10px")
    )

    display(grid)
```

```python
import torch
#use this for models base on SDXL
from diffusers import StableDiffusionXLPipeline
#use this for models not base on SDXL
#from diffusers import StableDiffusionPipeline

pipe = StableDiffusionXLPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0",torch_dtype=torch.float16)
pipe = pipe.to("cuda")
pipe.safety_checker = None
```

# **Implement**

```python
prompt = "(masterpiece:1.2),(amazing quality),cyberpunk style,"
h=1360
w=1024
steps=25
guidance=7
num=3
neg = "(bad quality:1.2),(worst quality:1.2),bad anatomy,simply background,jpeg artifacts,ugly,poorly drawn,blurrywatermark,simple background,(extrahands:1.2),emoji,sketch,draft,conjoined fingers, deformed fingers, ugly eyes, imperfect eyes,(extra_fingers:1.2)"
image = []

for i in range(num):
  result = pipe(prompt, height=h, width=w, num_inference_steps=steps, guidance_scale=guidance, negative_prompt=neg).images[0]
  image.append(result)
interactive_image_downloader(image)
```
